Q-3. Consider the ‘Blood Pressure Before’ and ‘Blood Pressure After’ columns from the data and calculate the following

*	Measure the dispersion in both and interpret the results.
*	Calculate mean and 5% confidence interval and plot it in a graph
*	Calculate the Mean absolute deviation and Standard deviation and interpret the results.
*	Calculate the correlation coefficient and check the significance of it at 1% level of significance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('data.csv')

# Calculate the dispersion in 'Blood Pressure Before' and 'Blood Pressure After'
dispersion_before = np.std(data.iloc[:, 1])
dispersion_after = np.std(data.iloc[:, 2])
print("Dispersion in 'Blood Pressure Before':", dispersion_before)
print("Dispersion in 'Blood Pressure After':", dispersion_after)

# Calculate the mean and 5% confidence interval
mean_before = np.mean(data.iloc[:, 1])
mean_after = np.mean(data.iloc[:, 2])
confidence_interval = stats.t.interval(0.95, len(data)-1, loc=mean_before, scale=stats.sem(data.iloc[:, 1]))
print("Mean 'Blood Pressure Before':", mean_before)
print("5% Confidence Interval for 'Blood Pressure Before':", confidence_interval)

# Plot the mean and confidence interval
plt.errorbar(1, mean_before, yerr=(confidence_interval[1]-mean_before), fmt='o', label='Blood Pressure Before')
plt.plot(2, mean_after, 'ro', label='Blood Pressure After')
plt.xticks([1, 2], ['Before', 'After'])
plt.ylabel('Mean Blood Pressure (mmHg)')
plt.title('Mean and Confidence Interval')
plt.legend()
plt.show()

# Calculate Mean Absolute Deviation (MAD) and Standard Deviation (SD)
mad_before = np.mean(np.abs(data.iloc[:, 1] - mean_before))
sd_before = np.std(data.iloc[:, 1])
print("Mean Absolute Deviation (MAD) of 'Blood Pressure Before':", mad_before)
print("Standard Deviation (SD) of 'Blood Pressure Before':", sd_before)

# Calculate the correlation coefficient and check significance
correlation_coefficient, p_value = stats.pearsonr(data.iloc[:, 1], data.iloc[:, 2])
print("Correlation Coefficient:", correlation_coefficient)
print("Significance at 1% level:", p_value < 0.01)
```
