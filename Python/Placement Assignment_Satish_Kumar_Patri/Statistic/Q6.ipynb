```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.read_csv('data.csv')

before = data.iloc[:, 1]
after = data.iloc[:, 2]

data['Change'] = after - before

plt.hist(data['Change'], bins=10)
plt.xlabel('Change in Blood Pressure')
plt.ylabel('Frequency')
plt.title('Distribution of Change in Blood Pressure')
plt.show()

stat, p_value = stats.shapiro(data['Change'])
alpha = 0.05

print(f"Test statistic: {stat}")
print(f"P-value: {p_value}")

if p_value > alpha:
    print("The change in blood pressure follows a normal distribution.")
else:
    print("The change in blood pressure does not follow a normal distribution.")
```
